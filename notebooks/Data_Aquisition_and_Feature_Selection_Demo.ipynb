```python
df_sub = df[["Sale Price","Waterfront View"]]
```

```python
df_sub_water_yes = df_sub[df_sub["Waterfront View"]=="Yes"]
df_sub_water_no = df_sub[df_sub["Waterfront View"]=="No"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
sns.kdeplot(data=df_sub_water_no['Sale Price'], label="Waterfront View No", color="blue")
sns.kdeplot(data=df_sub_water_yes['Sale Price'], label="Waterfront View Yes", color="red")
plt.xlabel("Sale Price")
plt.ylabel("Density")
plt.title("KDE Plot of Sale Price by Waterfront View")
plt.legend()
plt.show()
```
